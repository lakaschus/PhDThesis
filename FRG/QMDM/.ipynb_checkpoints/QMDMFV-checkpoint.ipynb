```python
def func(t, ux, uxx, uy, uyy, uxy, x, y, T, mu):
  k = np.exp(-t)*cutoffr
  sqEpi = k**2 + 2*ux 
  sqEs = sqEpi + 4*x*uxx
  sqENG = k**2 + 2*uy
  sqEG = sqENG + 4*y*uyy
  alpha0 = 16*mu**4 - 4*mu**2*sqEG - 4*mu**2*sqENG + sqEG*sqENG - 8*mu**2*sqEs + sqEG*sqEs + sqENG*sqEs - 16*uxy**2*x*y + 0.j
  alpha1 = 2*sqEG + 2*sqENG + 2*sqEs + 0.j
  alpha2 = 3 + 0.j
  beta0 = (-4*mu**2 + sqENG)*(-4*mu**2*sqEs + sqEG*sqEs - 16*uxy**2*x*y) + 0.j
  beta1 = 16*mu**4 - 4*mu**2*(sqEG + sqENG - 2*sqEs) + sqENG*sqEs + sqEG*(sqENG + sqEs) - 16*uxy**2*x*y + 0.j
  beta2 = 8*mu**2 + sqEG + sqENG + sqEs + 0.j
  beta3 = 1 + 0.j

  var1 = ( ( 27 * ( beta0 )**( 2 ) + ( 4 * ( beta1 )**( 3 ) + ( -18 * beta0 * beta1 * beta2 + ( -1 * ( beta1 )**( 2 ) * ( beta2 )**( 2 ) + 4 * beta0 * ( beta2 )**( 3 ) ) ) ) ) )**( 1/2 )
  var2 = ( -27 * beta0 + ( 9 * beta1 * beta2 + -2 * ( beta2 )**( 3 ) ) )

  z1 = ( ( 6 )**( -1/2 ) * ( ( ( 3 * ( 3 )**( 1/2 ) * var1 + var2 ) )**( -1/3 ) * ( ( ( 6 * ( 3 )**( 1/2 ) * var1 + 2 * var2 ) )**( 2/3 ) + ( -6 * ( 2 )**( 1/3 ) * beta1 + ( -2 * ( ( 3 * ( 3 )**( 1/2 ) * var1 + var2 ) )**( 1/3 ) * beta2 + 2 * ( 2 )**( 1/3 ) * ( beta2 )**( 2 ) ) ) ) )**( 1/2 ) )
  z2 = (1/2 * ( 3 )**( -1/2 ) * ( ( ( 3 * ( 3 )**( 1/2 ) * var1 + var2 ) )**( -1/3 ) * ( complex( 0,1 ) * ( complex( 0,1 ) + ( 3 )**( 1/2 ) ) * ( ( 6 * ( 3 )**( 1/2 ) * var1 + 2 * var2 ) )**( 2/3 ) + ( 6 * ( 2 )**( 1/3 ) * ( 1 + complex( 0,1 ) * ( 3 )**( 1/2 ) ) * beta1 + ( -4 * ( ( 3 * ( 3 )**( 1/2 ) * var1 + var2 ) )**( 1/3 ) * beta2 + complex( 0,-2 ) * ( 2 )**( 1/3 ) * ( complex( 0,-1 ) + ( 3 )**( 1/2 ) ) * ( beta2 )**( 2 ) ) ) ) )**( 1/2 ))
  z3 = (1/2 * ( 3 )**( -1/2 ) * ( ( ( 3 * ( 3 )**( 1/2 ) * var1 + var2 ) )**( -1/3 ) * ( ( -1 + complex( 0,-1 ) * ( 3 )**( 1/2 ) ) * ( ( 6 * ( 3 )**( 1/2 ) * var1 + 2 * var2 ) )**( 2/3 ) + ( 6 * ( 2 )**( 1/3 ) * ( 1 + complex( 0,-1 ) * ( 3 )**( 1/2 ) ) * beta1 + ( -4 * ( ( 3 * ( 3 )**( 1/2 ) * var1 + var2 ) )**( 1/3 ) * beta2 + complex( 0,2 ) * ( 2 )**( 1/3 ) * ( complex( 0,1 ) + ( 3 )**( 1/2 ) ) * ( beta2 )**( 2 ) ) ) ) )**( 1/2 ))

  res = ( -1/2 * ( z1 )**( -1 ) * ( ( -1 * ( z1 )**( 2 ) + ( z2 )**( 2 ) ) )**( -1 ) * ( ( -1 * ( z1 )**( 2 ) + ( z3 )**( 2 ) ) )**( -1 ) * ( alpha0 + ( ( z1 )**( 2 ) * alpha1 + ( z1 )**( 4 ) * alpha2 ) ) * 
         1./np.tan( 1/2 * ( T )**( -1 ) * z1 ) + ( -1/2 * ( z2 )**( -1 ) * ( ( ( z1 )**( 2 ) + -1 * ( z2 )**( 2 ) ) )**( -1 ) * ( ( -1 * ( z2 )**( 2 ) + ( z3 )**( 2 ) ) )**( -1 ) * ( alpha0 + ( ( z2 )**( 2 ) * alpha1 + ( z2 )**( 4 ) * alpha2 ) ) * 1./np.tan( 1/2 * ( T )**( -1 ) * z2 ) + -1/2 * ( z3 )**( -1 ) * ( ( ( z1 )**( 2 ) + -1 * ( z3 )**( 2 ) ) )**( -1 ) * ( ( ( z2 )**( 2 ) + -1 * ( z3 )**( 2 ) ) )**( -1 ) * ( alpha0 + ( ( z3 )**( 2 ) * alpha1 + ( z3 )**( 4 ) * alpha2 ) ) * 1./np.tan( 1/2 * ( T )**( -1 ) * z3 ) ) )
  
  return np.real(res)

def Epion(u, t):
  return np.sqrt(np.exp(-2*t)*cutoffr**2 + 2*u)

def Epsi(x, y, t, mu, n):
  return np.sqrt(hy**2*y + ( np.sqrt(np.exp(-2*t)*cutoffr**2 + hx**2*x) + (-1)**n*mu )**2)

def ENG(uy, t):
  return np.sqrt(np.exp(-2*t)*cutoffr**2 + 2*uy)

def src(x, y, t, T, mu):
  Eps1 = Epsi(x, y, t, mu, 0)
  Eps2 = Epsi(x, y, t, mu, 1)
  eps = np.sqrt(np.exp(-2*t)*cutoffr**2 + hx**2*x)
  return 12/(12*np.pi**2)*np.exp(-5*t)*cutoffr**5*(1/Eps1*(1 + mu/eps)*(1 - 2/(1 + np.exp(Eps1/T))) \
                               +1/Eps2*(1 - mu/eps)*(1 - 2/(1 + np.exp(Eps2/T)))) 

def src_x(x, y, t, T, mu):
  Ekp = Epsi(x, y, t, mu, 0)
  Ekm = Epsi(x, y, t, mu, 1)
  eps = np.sqrt(np.exp(-2*t)*cutoffr**2 + hx**2*x)
  argp = Ekp/(2*T)
  argm = Ekm/(2*T)
  expr = 2*1/(12*np.pi**2)*np.exp(-5*t)*cutoffr**5*hx**2*( 1/(np.cosh(argp))*1/(T*Ekp**2*eps**3)*(- T*mu*Ekp*np.sinh(2*argp) \
                                                          + (Ekp - T*np.sinh(2*argp))*eps*(mu + eps)**2/Ekp )/np.cosh(argp) \
                                                        + 1/(np.cosh(argm))*1/(T*Ekm**2*eps**3)*( T*mu*Ekm*np.sinh(2*argm) \
                                                          + (Ekm - T*np.sinh(2*argm))*eps*(mu - eps)**2/Ekm )/np.cosh(argm) )
  Eps = eps
  expr2 = 1/(12*np.pi**2)*cutoffr**5*np.exp(-5*t)*hx**2/(Eps**3*T)*( Eps*( 1/(np.cosh((Eps - mu)/(2*T))**2 + 1/(np.cosh((Eps + mu)/(2*T))**2 )) ) - 2*T*\
                                                               (np.tanh((Eps - mu)/(2*T)) + np.tanh((Eps + mu)/(2*T))))
  return expr+expr2
  
def src_y(x, y, t, T, mu):
  Ekp = Epsi(x, y, t, mu, 0)
  Ekm = Epsi(x, y, t, mu, 1)
  eps = np.sqrt(np.exp(-2*t)*cutoffr**2 + hx**2*x)
  argp = Ekp/(2*T)
  argm = Ekm/(2*T)
  expr = 1/(12*np.pi**2)*np.exp(-5*t)*cutoffr**5*hy**2/eps*(2/np.cosh(argm)*(-mu + eps)/np.cosh(argm)/(T*Ekm**2) - 4*np.tanh(argm)*(-mu + eps)/(Ekm**3) \
                                                          +2/np.cosh(argp)*(mu + eps)/np.cosh(argp)/(T*Ekp**2) - 4*np.tanh(argp)*(mu + eps)/(Ekp**3))
  return expr

# convection flux
def flux(v, w, t, T, mu):
  Ep = Epion(v, t)
  f = 3/(12*np.pi**2)*np.exp(-5*t)*cutoffr**5*1/(np.tanh( Ep/(2*T) ))/Ep
  Eng = ENG(w, t)
  g = 2/(6*np.pi**2)*np.exp(-5*t)*cutoffr**5*1/(Eng)*(np.sinh(Eng/T)/(np.cosh(Eng/T) - np.cosh(2*mu/T)))
  return (f+g, np.zeros(f.shape))

# characteristic velocity/local propagation speed
def eigenvalues_f(v, t, T, mu):
  Ep = Epion(v, t)
  ev1 = (-1/(8*np.pi**2*Ep**3*T)*np.exp(-5*t)*cutoffr**5/( np.sinh(Ep/(2*T)) )*( Ep + T*np.sinh(Ep/T) )/(np.sinh(Ep/(2*T))))
  ev2 = np.zeros(v.shape)
  ev_max = np.maximum(ev1, ev2)
  ev_min = np.minimum(ev1, ev2)
  return ( (ev_max > np.zeros(ev_max.shape)).astype(int)*ev_max, (ev_min < np.zeros(ev_min.shape)).astype(int)*ev_min )

# convection flux
def glux(v, w, t, T, mu):
  Ep = Epion(v, t)
  f = 3/(12*np.pi**2)*np.exp(-5*t)*cutoffr**5*1/(np.tanh( Ep/(2*T) ))/Ep
  Eng = ENG(w, t)
  g = 2/(6*np.pi**2)*np.exp(-5*t)*cutoffr**5*1/(Eng)*(np.sinh(Eng/T)/(np.cosh(Eng/T) - np.cosh(2*mu/T)))
  return (np.zeros(f.shape), +f+g)

# characteristic velocity/local propagation speed
def eigenvalues_g(w, t, T, mu):
  Eng = ENG(w, t)
  ev1 = (1/(6*np.pi**2)*np.exp(-5*t)*cutoffr**5*( 2*Eng + 2*np.cosh(2*mu/T)*(-Eng*np.cosh(Eng/T) + T*np.sinh(Eng/T)) - T*np.sinh(2*Eng/T) )/
       (Eng**3*T)/(np.cosh(Eng/T) - np.cosh((2*mu)/T))**2)
  ev2 = np.zeros(w.shape)
  ev_max = np.maximum(ev1, ev2)
  ev_min = np.minimum(ev1, ev2)
  return ( (ev_max > np.zeros(ev_max.shape)).astype(int)*ev_max, (ev_min < np.zeros(ev_min.shape)).astype(int)*ev_min )

def diffx(t, v, vx, w, wy, uxy, x, y, T, mu):
  #Es = Esig(x, ux, uxx, t)
  return -1/(6*np.pi**2)*np.exp(-5*t)*cutoffr**5*func(t, v, vx, w, wy, uxy, x, y, T, mu) #+ src(x, y, t, T, mu)#!

def diffy(t, w, wy, ux, uxx, uxy, x, y, T, mu):
  #Es = Esig(x, ux, uxx, t)
  return -1/(6*np.pi**2)*np.exp(-5*t)*cutoffr**5*func(t, ux, uxx, w, wy, uxy, x, y, T, mu) #+ src(x, y, t, T, mu) #- flux(x, y, ux, t, T, mu) #!

def construct_potential_x(x, u, Ny):
    v = np.zeros(u.shape)
    for i in range(Ny):
      v[i,:] = integrate.cumtrapz(u[i,:], x[i,:], initial=0)
    return v

def construct_potential_y(y, u, Nx):
    v = np.zeros(u.shape)
    for i in range(Nx):
      v[:,i] = integrate.cumtrapz(u[:,i], y[:,i], initial=0)
    return v

def construct_potential_full(u, v, Nx):
  return u + np.tile(v[:,0], (Nx, 1)).T

def construct_potential_full_2(u, v, Ny):
  return v + np.tile(u[0,:], (Ny, 1))

def minmod(a, b, c):
  th = 1.0
  maximum = np.maximum(th*a, np.maximum(b, th*c))
  minimum = np.minimum(th*a, np.minimum(b, th*c))
  hs_pos = np.heaviside(a, 1)*np.heaviside(b, 1)*np.heaviside(c, 1)
  hs_neg = np.heaviside(-a, 1)*np.heaviside(-b, 1)*np.heaviside(-c, 1)
  return minimum*hs_pos + maximum*hs_neg

def construct_boundaries(u):
  ul_bnd = u[:,0] - (u[:,1] - u[:,0]) #3*u[:,0] - 3*u[:,1] + u[:,2] #4*u[:,0] - 6*u[:,1] + 4*u[:,2] - 1*u[:,3]#
  ur_bnd = u[:,-1] + (u[:,-1] - u[:,-2])#3*u[:,-1] - 3*u[:,-2] + u[:,-3]  #4*u[:,-1] - 6*u[:,-2] + 4*u[:,-3] - 1*u[:,-4] #5*u[:,-1] - 10*u[:,-2] + 10*u[:,-3] - 5*u[:,-4] + u[:,-5]  #

  u = np.concatenate((np.array([ul_bnd]).T, np.concatenate((u, np.array([ur_bnd]).T), axis=1)), axis=1)

  ut_bnd = u[0,:] - (u[1,:] - u[0,:])#3*u[0,:] - 3*u[1,:] + u[2,:]#4*u[0,:] - 6*u[1,:] + 4*u[2,:] - 1*u[3,:] #5*u[0,:] - 10*u[1,:] + 10*u[2,:] - 5*u[3,:] + u[4,:] ##
  ub_bnd = u[-1,:] + (u[-1,:] - u[-2,:])#3*u[-1,:] - 3*u[-2,:] + u[-3,:] #4*u[-1,:] - 6*u[-2,:] + 4*u[-3,:] - 1*u[-4,:] #5*u[-1,:] - 10*u[-2,:] + 10*u[-3,:] - 5*u[-4,:] + u[-5,:]##3*u[-1,:] - 3*u[-2,:] + u[-3,:] 

  u = np.concatenate((np.array([ut_bnd]), np.concatenate((u, np.array([ub_bnd])), axis=0)), axis=0)
  return u

def construct_u(u):
  xlimiter = minmod((u[1:-1, 1:-1] - u[1:-1,:-2])/dx, (u[1:-1,2:] - u[1:-1,:-2])/(2*dx), (u[1:-1,2:] - u[1:-1,1:-1])/dx)
  uE = u[1:-1,1:-1] + dx/2*xlimiter
  uW = u[1:-1,1:-1] - dx/2*xlimiter

  ylimiter = minmod((u[1:-1,1:-1] - u[:-2,1:-1])/dy, (u[2:,1:-1] - u[:-2,1:-1])/(2*dy), (u[2:,1:-1] - u[1:-1,1:-1])/dy) 
  uN = u[1:-1,1:-1] + dy/2*ylimiter
  uS = u[1:-1,1:-1] - dy/2*ylimiter

  uNE = uE + dy/2*ylimiter
  uNW = uW + dy/2*ylimiter
  uSE = uE - dy/2*ylimiter
  uSW = uW - dy/2*ylimiter 
  return (u[1:-1,1:-1], xlimiter, ylimiter, uE, uW, uN, uS, uNE, uNW, uSE, uSW)

def f_interf(vNE, vNW, vSE, vSW, wNE, wNW, wSE, wSW, a_plus, a_minus, t, T, mu):
  fluxNE, fluxNW = flux(vNE, wNE, t, T, mu), flux(vNW, wNW, t, T, mu)
  fluxSE, fluxSW = flux(vSE, wSE, t, T, mu), flux(vSW, wSW, t, T, mu)
  den = 2*(a_plus - a_minus)
  pre_fac = a_plus*a_minus/den
  v = a_plus*(fluxNE[0] + fluxSE[0])/den - a_minus*(fluxNW[0] + fluxSW[0])/den + pre_fac*(vNW - vNE + vSW - vSE)
  w = a_plus*(fluxNE[1] + fluxSE[1])/den - a_minus*(fluxNW[1] + fluxSW[1])/den + pre_fac*(wNW - wNE + wSW - wSE)
  return (v, w)

def g_interf(vNE, vNW, vSE, vSW, wNE, wNW, wSE, wSW, a_plus, a_minus, t, T, mu):
  fluxNE, fluxNW = glux(vNE, wNE, t, T, mu), glux(vNW, wNW, t, T, mu)
  fluxSE, fluxSW = glux(vSE, wSE, t, T, mu), glux(vSW, wSW, t, T, mu)
  
  den = 2*(a_plus - a_minus)
  pre_fac = a_plus*a_minus/den
  v = a_plus*(fluxNW[0] + fluxNE[0])/den - a_minus*(fluxSW[0] + fluxSE[0])/den + pre_fac*(vSW - vNW + vSE - vNE)
  w = a_plus*(fluxNW[1] + fluxNE[1])/den - a_minus*(fluxSW[1] + fluxSE[1])/den + pre_fac*(wSW - wNW + wSE - wNE)
  return (v, w)
```

```python
def dudt(u, t, T, mu):
  #print("flow scale (k/cutoff): ", cutoffr*np.exp(-t))
  global t_stop,flag
  v, w = u[:int(len(u)/2)].reshape(-1, Nx), u[int(len(u)/2):].reshape(-1, Nx)

  v, w = construct_boundaries(construct_boundaries(v)), construct_boundaries(construct_boundaries(w))

  v, vx_lim, vy_lim, vE, vW, vN, vS, vNE, vNW, vSE, vSW = construct_u(v)
  w, wx_lim, wy_lim, wE, wW, wN, wS, wNE, wNW, wSE, wSW = construct_u(w)

  evL_max, evL_min = eigenvalues_f(vW[:,1:], t, T, mu)
  evR_max, evR_min = eigenvalues_f(vE[:,:-1], t, T, mu)

  comp = (evL_max >= evR_max).astype(int)
  a_plus = comp*evL_max + (1 - comp)*evR_max

  comp = (evL_min < evR_min).astype(int)
  a_minus = comp*evL_min + (1 - comp)*evR_min
 
  evN_max, evN_min = eigenvalues_g(wN[:-1,:], t, T, mu)
  evS_max, evS_min = eigenvalues_g(wS[1:,:], t, T, mu)

  comp = (evN_max >= evS_max).astype(int)
  b_plus = comp*evN_max + (1 - comp)*evS_max

  comp = (evS_min < evN_min).astype(int)
  b_minus = comp*evS_min + (1 - comp)*evN_min

  fi = f_interf(vNE[:,:-1], vNW[:,1:], vSE[:,:-1], vSW[:,1:], wNE[:,:-1], wNW[:,1:], wSE[:,:-1], wSW[:,1:], \
               a_plus, a_minus, t, T, mu)
  v_fif, w_fif = fi[0], fi[1]

  gi = g_interf(vNE[:-1,:], vNW[:-1,:], vSE[1:,:], vSW[1:,:], wNE[:-1,:], wNW[:-1,:], wSE[1:,:], wSW[1:,:],\
                b_plus, b_minus, t, T, mu)
  v_gif, w_gif = gi[0], gi[1]

  # Diffusion part #################
  v_x = (v[1:-1, 2:] - v[1:-1, 1:-1])/dx
  vx = d_dx(v) #(v[1:-1, 2:] - v[1:-1, 1:-1])/dx
  wx = d_dx(w)
  wy = d_dy(w)
  vy = d_dy(v)
  uxy = (vy+wx)/2

  Pxp = 1/2*(diffx(t, v[1:-1, 1:-1], v_x, w[1:-1, 1:-1], wy[1:-1, 1:-1], uxy[1:-1, 1:-1], xcgrd_R[:-1,:-1], ycgrd, T, mu) \
           + diffx(t, v[1:-1, 2:], v_x, w[1:-1, 2:], wy[1:-1, 2:], uxy[1:-1, 2:], xcgrd_R[:-1, 1:], ycgrd, T, mu))

  v_x = (v[1:-1, 1:-1] - v[1:-1, :-2])/dx

  Pxm = 1/2*(diffx(t, v[1:-1, :-2], v_x, w[1:-1, :-2], wy[1:-1, :-2], uxy[1:-1, :-2], xcgrd_L[:-1,:-1], ycgrd, T, mu) \
           + diffx(t, v[1:-1, 1:-1], v_x, w[1:-1, 1:-1], wy[1:-1, 1:-1], uxy[1:-1, 1:-1], xcgrd_L[:-1, 1:], ycgrd, T, mu))

  w_y = (w[2:,1:-1] - w[1:-1, 1:-1])/dy

  Pyp = 1/2*(diffy(t, w[1:-1, 1:-1], w_y , v[1:-1, 1:-1], vx[1:-1, 1:-1], uxy[1:-1, 1:-1], xcgrd, ycgrd_R[:-1, :-1], T, mu) \
           + diffy(t, w[2:, 1:-1], w_y , v[2:, 1:-1], vx[2:, 1:-1], uxy[2:, 1:-1], xcgrd, ycgrd_R[1:, :-1], T, mu))

  w_y = (w[1:-1,1:-1] - w[:-2, 1:-1])/dy

  Pym = 1/2*(diffy(t, w[:-2, 1:-1], w_y , v[:-2, 1:-1], vx[:-2, 1:-1], uxy[:-2, 1:-1], xcgrd, ycgrd_L[:-1, :-1], T, mu) \
           + diffy(t, w[1:-1, 1:-1], w_y , v[1:-1, 1:-1], vx[1:-1, 1:-1], uxy[1:-1, 1:-1], xcgrd, ycgrd_L[1:, :-1], T, mu))

  dvdt =  (v_fif[1:-1, :-1] - v_fif[1:-1, 1:])/dx + (v_gif[:-1, 1:-1] - v_gif[1:, 1:-1])/dx  + (Pxp - Pxm)/dx + src_x(xcgrd, ycgrd, t, T, mu) 
  dwdt =  ((w_fif[1:-1, :-1] - w_fif[1:-1, 1:])/dy + (w_gif[:-1, 1:-1] - w_gif[1:, 1:-1])/dy) + (Pyp - Pym)/dy + src_y(xcgrd, ycgrd, t, T, mu)

  dVdt = construct_potential_x(xcgrd, dvdt, Ny)
  dWdt = construct_potential_y(ycgrd, dwdt, Nx)
  dVWdt = construct_potential_full(dVdt, dWdt, Nx) #+ construct_potential_full_2(V, W, Nx))/2
  dvdt, dwdt = d_dx(dVWdt), d_dy(dVWdt)
    
  rhs = np.append( np.squeeze(np.asarray(dvdt)).flatten(), np.squeeze(np.asarray(dwdt)).flatten() )

  if t >= t_stop:
      print("\n")
      print("flow scale (k/cutoff): ", cutoffr*np.exp(-t))
      t_stop += 0.1
  return rhs
```

```python
def solve(T, mu):
    v0x_vec = np.squeeze(np.asarray(v0x(xcgrd, ycgrd))).flatten()
    v0y_vec = np.squeeze(np.asarray(v0y(xcgrd, ycgrd))).flatten()
    u_init = np.append(v0x_vec, v0y_vec) # v = w = u0
    return solve_ivp(lambda t, y: dudt(y, t, T, mu), (t[0], t[-1]), u_init, method='RK45', rtol=1e-11, atol=1e-11)
```

Minimum:

```python
expl = c*np.sqrt(xcgrd) + 2*muV**2*ycgrd
argm = np.unravel_index(np.argmin(V-expl, axis=None), V.shape)
fpi = np.sqrt(argm[1]*dx)
diq = np.sqrt(argm[0]*dy)
print("argmin: ", argm)
print("fpi: ", fpi*cutoff)
print("diq: ", diq*cutoff)
```
